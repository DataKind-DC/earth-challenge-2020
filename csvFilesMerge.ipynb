```python
import pandas as pd
pd.set_option("display.max_columns", None)
```

```python
marine = pd.read_csv('MarineLitterWatch_totals_by_category.csv')
```

```python
tides = pd.read_csv('TIDES_earth_totals_by_category.csv')
```

```python
mdmap = pd.read_csv('mdmap_totals_by_category.csv')
```

```python
marine['source'] = 'marine'
mdmap['source'] = 'mdmap'
tides['source'] = 'tides'
```

```python
print(mdmap.shape)
print(tides.shape)
print(marine.shape)
```

```python
#preparing the marine data
```

```python
marine.head(5)
```

```python
#added 60,000 to uniqueID to make sure all the values from different files are unique.
marine['UniqueID'] = marine['UniqueID'] + 60000
```

```python
#marList has all the column names in Marine data
marList = list(marine.columns.values.tolist())
marList.sort()
```

```python
#funtion to set the format of the date in marine dataset
def marDate(s):
    s = str(s)
    return(s[4:6]+'-'+s[6:] +'-'+s[0:4])
```

```python
marine.Date= marine.Date.apply(marDate)
```

```python
#preparing the tides data
```

```python
tides.head(5)
```

```python
#renaming the Cleanup Date to Date in tides dataset
tides = tides.rename(columns = {'Cleanup Date': 'Date'})
```

```python
#splitting the Date column to get month, year and Date in different column and then joining it back to the dataframe.
tidesTemp = tides['Date'].str.split('-', expand = True)
tides = tides.join(tidesTemp)
```

```python
#renaming the columns to make sure they are consistant in all the 3 datasets. 
tides = tides.rename(columns = {'Cleanup ID': 'UniqueID',
                                'Cleanup Date': 'Date',
                                'People': 'Total Number of Volunteers',
                                'Total Items Collected': 'Total Item Count',
                                'LAT': 'Latitude 1',
                                'LONG': 'Longitude 1',
                                'Fishing Gear': 'Fishing Gear Raw Count',
                                'Glass': 'Glass Raw Count',
                                'Metal': 'Metal Raw Count',
                                'Other': 'Other Raw Count',
                                'Plastic': 'Plastic Raw Count',
                                'Processed Lumber':'Processed Lumber Raw Count',
                                'Processed Lumber Raw Count': 'Processed Lumbar Raw Count',
                                'Rubber': 'Rubber Raw Count',
                                'source': 'Dataset',
                               0: 'Year',
                               1: 'Month',
                               2: 'Day'})
```

```python
#dropping the GPS column in tides Data
tides.drop('GPS', axis=1, inplace=True)
```

```python
# to get the list of all the column names in tides
tidList =  list(tides.columns.values.tolist())
tidList.sort()
```

```python
#function to set date format in tides data
def tidDate(s):
    s =str(s)
    return(s[5:7]+'-'+s[-2:]+ '-'+ s[0:4])
```

```python
tides.Date = tides.Date.apply(tidDate)
```

```python
#preping the data for mdmap
```

```python
mdmap.head(5)
```

```python
#splitting the Date column to get month, year and Date in different column and then joining it back to the dataframe. 
mdmapTemp = mdmap['Date'].str.split('-', expand = True)
mdmap = mdmap.join(mdmapTemp)
```

```python
#renaming the columns to make sure they are consistant in all the 3 datasets. 
mdmap = mdmap.rename(columns= {'Latitude Start': 'Latitude 1',
                               'Latitude End': 'Latitude 2',
                               'Longitude Start': 'Longitude 1',
                               'Longitude End': 'Longitude 2',
                               'Total Debris': 'Total Item Count',
                               'Cloth': 'Cloth Raw Count',
                               'Fishing Gear': 'Fishing Gear Raw Count',
                               'Glass': 'Glass Raw Count',
                               'Metal': 'Metal Raw Count',
                               'Other': 'Other Raw Count',
                               'Plastic': 'Plastic Raw Count',
                               'Processed Lumber': 'Processed Lumbar Raw Count',
                               'Processed Lumber Raw Count':'Processed Lumbar Raw Count',
                               'Processed Lumber Raw Count': 'Processed Lumbar Raw Count',
                               'Rubber':  'Rubber Raw Count',
                               0: 'Month',
                               1: 'Day',
                               2: 'Year',
                               'UniqueId': 'UniqueID',
                               'TotalArea': 'Total Area'})
```

```python
# to get the list of all the column names in tides
mdList = list(mdmap.columns.values.tolist())
mdList.sort()
```

```python
#combining the 3 dataframes into dataframe named frames. 
frames = [marine, tides, mdmap]
result = pd.concat(frames)
```

```python
result.shape
```

```python
result.head(5)
```

```python
#list of columns to be removed from the results dataframe. 
toRemoveList = ['# of bags', 'Adults','Children','Cloth Per Sq Meter','Fishing Gear  Per Sq Meter',
'Glass Per Sq Meter','Group Name','Length','Metal Per Sq Meter','Meters','Miles','Organization',
'Other Per Sq Meter','Plastic Count','Plastic Per Sq Meter','Pounds','Processed Lumber Per Sq Meter',
'Rubber Per Sq Meter','Survey ID','Survey Year','Total Debris Per Sq Meter','Width','Zone']
len(toRemoveList)
```

```python
#dropping all the column in the toRemoveList
result.drop(toRemoveList, axis=1, inplace=True)
```

```python
result["Dataset"].fillna("MDMAP", inplace = True) 
```

```python
set(result.Dataset)
```

```python
#filling NaN in result Df with value 'MDMAP' - only MDMAP has NULL values in Dataset column
result.to_csv('concatinated2.csv', sep=',')
```
